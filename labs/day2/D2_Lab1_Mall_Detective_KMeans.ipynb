### 💪 Section 2: The Elbow Method — Let the Data Choose K

Remember from theory: we try K = 1 to 10, measure **inertia** (how tight the clusters are — lower is tighter), and look for the **elbow** where improvement suddenly slows down.

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

inertias = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)
    inertias.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertias, marker='o', color='slateblue')
plt.title('💪 The Elbow Method')
plt.xlabel('K (number of clusters)')
plt.ylabel('Inertia (lower = tighter clusters)')
plt.xticks(range(1, 11))
plt.grid(alpha=0.3)
plt.show()

# 👀 Where does the curve BEND? That's your K!
```

### 🤖 Section 3: Run K-Means & Reveal the Segments!

```python
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(9, 7))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'],
            c=df['Cluster'], cmap='tab10', alpha=0.7)

# Plot the centroids as big X marks ❌
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=250, marker='X', label='Centroids')

plt.title(f'🕵️ Case Solved: {best_k} Customer Segments Discovered!')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
# Detective report: the profile of each segment 📋
report = df.groupby('Cluster')[['Annual Income (k$)', 'Spending Score (1-100)', 'Age']].mean().round(1)
report['Customers'] = df['Cluster'].value_counts().sort_index()
report
```
