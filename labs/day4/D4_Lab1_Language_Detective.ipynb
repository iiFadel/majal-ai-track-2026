### ✂️ Section 1: Tokenization — Chop It Up

Step one of EVERY NLP system, from spam filters to ChatGPT: split text into **tokens**.

```python
sentence = "The students absolutely loved the AI initiative, didn't they?!"

tokens = word_tokenize(sentence)
print(f"✂️ {len(tokens)} tokens:")
print(tokens)

# 👀 Notice: "didn't" became ["did", "n't"], and punctuation gets its own tokens.
# Tokenization has opinions!
```

### 🧹 Section 2: Cleaning — Stopwords, Stems, Lemmas

**Stopwords** are ultra-common words ("the", "is", "of") that carry little meaning alone. Classic NLP drops them to focus on the words that matter.

```python
english_stopwords = set(stopwords.words('english'))
print(f"📋 English has ~{len(english_stopwords)} stopwords. A few: {sorted(list(english_stopwords))[:12]}")

kept = []

for tok in tokens:
    # check if token is not a stopword (case-insensitive)
    if tok.lower() not in english_stopwords:
        # check if token contains only alphabetic characters
        if tok.isalpha():
            kept.append(tok)

# Faster way: list comprehension does the same loop + two if-checks in one line.
# Read it right to left: "for tok in tokens" generates each token,
# then both conditions after "if" must be true for tok to be added to the list.
# It's called a "filter" comprehension since it's just building a new list
# from an old one, keeping only items that pass a test, no transformation applied.


# kept = [tok for tok in tokens if tok.lower() not in english_stopwords and tok.isalpha()]


print(f"\nBefore: {tokens}")
print(f"After:  {kept}   ← only the meaningful words survive")
```

```python
# Stemming (fast & dumb) vs Lemmatization (smart & dictionary-based)
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words = ['running', 'studies', 'better', 'was', 'mice', 'loving']

df = pd.DataFrame({
    'word': words,
    'stem  (chop the ending)': [stemmer.stem(w) for w in words],
    'lemma (dictionary root)': [lemmatizer.lemmatize(w, pos='v') for w in words],
})
df

# 👀 Look at 'studies' and 'was': the stemmer produces weird non-words,
# the lemmatizer finds the real root. Speed vs accuracy — a classic ML trade-off!
```

### 🎒 Section 3: Bag of Words — Just Count!

The oldest trick in NLP: represent a sentence as a **vector of word counts**. Ignore grammar, ignore order — just count. Let's see it on three movie reviews.

```python
reviews = [
    "I love love this movie",
    "I hate this movie",
    "great acting great story",
]

vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(reviews)

pd.DataFrame(bow.toarray(), columns=vectorizer.get_feature_names_out(), index=reviews)

# Every row is now a NUMBER VECTOR — exactly what Day 1's models eat!
# This + logistic regression = a working spam filter. Old school, still used.
```

### 💥 Section 4: Break It! — The Fatal Flaws

Time to be the villain. 😈 Bag of Words has two famous weaknesses. Let's expose both with four short sentences.

```python
# FLAW 1: word order is gone
pair = ["the dog bit the man", "the man bit the dog"]
v = CountVectorizer()
vecs = v.fit_transform(pair).toarray()

print(pd.DataFrame(vecs, columns=v.get_feature_names_out(), index=pair))
print(f"\n💥 Are the two vectors IDENTICAL? {(vecs[0] == vecs[1]).all()}")
print("To Bag of Words, these two very different events are THE SAME SENTENCE. 😬")
```

```python
# FLAW 2: no idea of meaning
pair2 = ["the film was great", "the movie was excellent"]
v2 = CountVectorizer()
vecs2 = v2.fit_transform(pair2).toarray()

print(pd.DataFrame(vecs2, columns=v2.get_feature_names_out(), index=pair2))

shared = (vecs2[0] * vecs2[1]).sum()
print(f"\n💥 Shared meaningful words between the two vectors: only their overlap on 'the'/'was'!")
print("These sentences mean the SAME thing — but 'film'/'movie' and 'great'/'excellent'")
print("are totally unrelated columns to a Bag of Words. It cannot see meaning. 😬")
```
